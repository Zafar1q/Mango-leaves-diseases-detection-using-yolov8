```python
# Define the path to the source folder and the destination folder
src_folder = '/content/gdrive/MyDrive/yolov8Fyp/FYPDTA1OG/prowdey'
dest_folder = '/content/gdrive/MyDrive/yolov3/SAug'

# Define the size for the resized images
width = 640
height = 640

# Loop through the source folder and resize each image
for filename in os.listdir(src_folder):
  # Read the image
  image = cv2.imread(os.path.join(src_folder, filename))
  
  # Resize the image
  resized_image = cv2.resize(image, (width, height))
  
  # Save the resized image in the destination folder
  cv2.imwrite(os.path.join(dest_folder, filename), resized_image)
       
```

```python
import cv2
import os

folder_path = '/content/gdrive/MyDrive/processing output'

for filename in os.listdir(folder_path):
    # load image
    img = cv2.imread(os.path.join(folder_path, filename))   # load image
    # denoise image
    img = cv2.fastNlMeansDenoisingColored(img, None, 5, 8, 7, 21) # color denoise image Mean filter
    # save denoised image
    cv2.imwrite(os.path.join(folder_path, '/content/gdrive/MyDrive/noiseimage' + filename), img) # save denoised image

```
