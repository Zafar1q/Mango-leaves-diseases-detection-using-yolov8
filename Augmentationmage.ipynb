```python


import tensorflow as tf #computer vision model k leya use hota hai
from tensorflow import keras  #computer vision model k leya use hota hai
from keras.preprocessing.image import ImageDataGenerator #image data augmentation generator library
from keras.preprocessing import image
import numpy as np
import matplotlib.pyplot as plt #graphical representation
import os
import PIL # image data library
```

```python
data_folder_path = '/content/Untitled Folder/'
augmented_folder_path = '/content/drive/MyDrive/audt'
```

```python
datagen = ImageDataGenerator(        #Data Augmentation Technique
        zoom_range=0.35,
        horizontal_flip=True,
        fill_mode='nearest')
```

```python
for filename in os.listdir(data_folder_path): #read folder
    img = PIL.Image.open(data_folder_path + filename) #open image dataset
    img = np.array(img)  #image convert into array
    img = img.reshape((1,) + img.shape)

    # Apply the augmentation techniques and save the augmented images
    i = 0
    for batch in datagen.flow(img, batch_size=20, save_to_dir=augmented_folder_path, save_prefix='aug_' + filename[:-1], save_format='jpeg'):
        i += 1
        if i > 2:
            break
```
